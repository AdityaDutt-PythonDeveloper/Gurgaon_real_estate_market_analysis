Problem Statement:

Suppose you are working as a Data Analyst for a real estate advisory firm operating in Gurgaon. The firm spents significant time to collect data on residential properties accross various localities in Gurgaon.

The firm wants data-backed insgihts to guide buyers, investors, and developers in making real estate decisions.

```python
# Step-1: import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step-2 : Load the Dataset
df = pd.read_csv('https://raw.githubusercontent.com/AdityaDutt-PythonDeveloper/Gurgaon_real_estate_market_analysis/refs/heads/main/data%20of%20gurugram%20real%20Estate.csv')
df.head()

# Allows us to inspect the structure of dataset
# Helps verify column names and data types
```

```python
# Step - 3: Basic dataset overview
df.shape #means we have 19515 rows and 12 columns
```

```python
df.info()
# Helps in identifying number of rows and columns
# reveals the missing values and incorrect data types
```

```python
# Data cleaning and preparation
# Step-4: Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.head()
```

```python
# debugging
df['price'].dtype
df['price'].head()
```

```python
# Step-5:convert numeric columns
df["price"] = df["price"].astype(str)
```

```python
df['price'] = df['price'].str.strip().str.replace(",", "")
```

```python
df['price'].astype(float)
```

```python
df['area'].astype(float).head()
```

```python
df['rate_per_sqft'].str.strip().str.replace(",", "").astype(float).head()
```

```python
# Step-6 : Clean Categorical Columns
df['status'] = df['status'].str.strip().str.lower()
df['status'].head()
```

```python
df['rera_approval'] = df['rera_approval'].str.strip().str.lower()
df['rera_approval'].head()
```

```python
df['flat_type'] = df['flat_type'].str.strip().str.lower()
df['flat_type'].head()
```

```python
# step- 7: remove duplicates
df = df.drop_duplicates()
df.shape  #duplicate rows can skew averages and insights
```

```python
df.shape
```

```python
# Question-1 : Which is the costiest flat?
temp = df['price'].idxmax()
df.loc[temp]
```

```python
# debugging
df['price'].dtype
# df['price'].astype(float)
```

```python
df["price"] = pd.to_numeric(
    df["price"].astype(str).str.replace(",", "", regex=False),
    errors="coerce"
)
```

```python
df["price"].dtype
```

```python
# Question-2 : which locality has highest average price
df.groupby("locality")["price"].mean().sort_values(ascending=False)
```

```python
df['rate_per_sqft'].dtype
```

```python
df["rate_per_sqft"] = pd.to_numeric(
    df["rate_per_sqft"].astype(str).str.replace(",", "", regex=False),
    errors="coerce"
)
```

```python
# Question-3 : Which locality has the highest rate per sq.ft?
df.groupby("locality")["rate_per_sqft"].mean().sort_values(ascending=False)
```

```python
# Question-4: Ready-to-move Vs Under-construction pricing
df.groupby(by='status')['price'].median()
```

```python
# Question-5: Does RERA approval affect pricing?
df.groupby('rera_approval')['price'].median()
```

Deaper Insights

```python
import matplotlib.pyplot as plt
```

```python
# Question-6: How does Area impact price?
df.plot(kind='scatter', x='area', y='price', title='Area vs Price', color='red', figsize=(8, 6))
plt.show()
```

```python
# Question - 7 : Which BHK config. is most expensive?
df.groupby('bhk_count')['price'].mean()
```

```python
# Question - 8:WHich property type is costliest?
df.groupby('flat_type')['price'].mean().sort_values(ascending=False)
```

```python
# Question - 9: Do certain builders price higher?
df.groupby(['company_name', 'flat_type'])['price'].mean().sort_values(ascending=False)
```

```python
# Question 10: Are larger homes more expensive per sqft.?
df.plot(kind='scatter', x='area', y='rate_per_sqft')
plt.show()
```
